## Feature Selection Techniques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "C:/Users/DEVENDRA/Desktop/ml_database_7/mobile_data.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

### Univariate feature

```python
x = df.iloc[:,:-1]
y = df['price_range']
```

```python
y.head()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
## apply selectkbest Algor
order_ranked_feat = SelectKBest(score_func=chi2, k=20 )
ordered_feat = order_ranked_feat.fit(x, y)
```

```python
ordered_score = pd.DataFrame(ordered_feat.scores_, columns=['score'])
ordered_colu = pd.DataFrame(x.columns)
```

```python
feature_imp = pd.concat([ordered_colu,ordered_score], axis = 1)
```

```python
feature_imp.columns=['feature', 'Score']
```

```python
feature_imp
```

```python
feature_imp.nlargest(10, 'Score')
```

### feature Importance
this technique gives a score for each feature of data , highest the score more relevant it is

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
rank_feat = pd.Series(model.feature_importances_, index= x.columns)
rank_feat.nlargest(10).plot(kind='barh')
```

```python
## find and remove correleation feature
threshold = 0.5
def correlation(df, threshold):
    col_corr = set() # set all the name of correlated columns
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i] # getting the name of the columns
                col_corr.add(colname)
    return col_corr
```

```python
correlation(df.iloc[:,:-1], threshold)
```
